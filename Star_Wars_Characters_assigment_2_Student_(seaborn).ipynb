Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">"TASK_1 Plotting the Star Wars characters by genders"</p>

style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
plt.rcParams["figure.figsize"] = (8,5)
```

```python
sw_gender=sw.gender.value_counts()
sw_gender
```

**This can now be plotted much more concisely.**<br>
Axes.bar_label() automatically labels bars [since matplotlib 3.4](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar_label.html)

```python
# With Seaborn Barplot
ax=sns.barplot(x=sw_gender.index, y=sw_gender.values)
plt.xticks(rotation=45)
plt.title('Gender', fontsize='13', color='b')
plt.xlabel('gender')
ax.bar_label(ax.containers[0]);
```

```python
# With Seaborn Countplot
ax=sns.countplot(data=sw, x='gender', order=sw_gender.index)
plt.xticks(rotation=45)
plt.title('Gender', fontsize='13', color='b')
ax.bar_label(ax.containers[0]);
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS under 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>


<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true"
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
#mass_cat
def categorize_mass(x):
    if x >= 100.0:
        return 'High_mass'
    elif 50.0 < x < 100.0:
        return 'Normal_mass'
    elif x <= 50.0:
        return 'Low_mass'
    else:
        return np.nan

sw['mass_cat'] = sw['mass'].transform(lambda x: categorize_mass(x))
sw_mass_cat=sw['mass_cat'].value_counts(dropna=False)
sw_mass_cat
```

```python
# With Seaborn Barplot
ax=sns.barplot(x=sw_mass_cat.index, y=sw_mass_cat.values)
plt.title('Mass Groups', fontsize='13', color='b')
plt.xlabel('mass_cat')
ax.bar_label(ax.containers[0]);
```

**Seaborn color palettes**

sns.color_palette("bone", n_colors=20)


https://seaborn.pydata.org/tutorial/color_palettes.html
https://www.practicalpythonfordatascience.com/ap_seaborn_palette

```python
# With Seaborn Countplot
ax=sns.countplot(data=sw, x='mass_cat', order=sw.mass_cat.value_counts(dropna=True).index, palette='flare')
plt.xticks(rotation=45)
plt.title('Mass Groups', fontsize='13', color='b')
ax.bar_label(ax.containers[0], fontsize=10, color='r');
```
